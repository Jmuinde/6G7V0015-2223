### Mapping the education levels

```python
ed_level = df2[(df2['education_level'] =='Total Ever Attended' ) | (df2['education_level'] == 'Total ever attended')|
(df2['education_level'] == 'All literate')].index
df2.drop(ed_level, inplace=True)
df2['education_level'].unique()
```

```python
ed_level1 = df2[(df2['education_level'] == 'Not Stated') | (df2['education_level'] == 'Not stated')].index
df2.drop(ed_level1, inplace=True)
df2['education_level'].unique()
```

```python
list1 = ['Never Attended', 'Informal', 'Preschool','Never attended','Pre- school', 'Non Formal', 'All illiterate', 'Non formal',  'General education',
         ]
list2 =['Primary Education', 'Grades 1-8',  'Secondary not completed', 'certificate not completed',
        'Grade 1 - 8', ]
list3=['Secondary Education', 'Secondary Completed','Pre- paratory', 'ten + 1 and ten + 2','Preparatory', 'Certificate', 'diploma or degree not completed',
       'Certificate Completed']
list4=['Above', 'Diploma Completed',  'degree and above completed', 'degree and above completed', 'Education completed',  'Diploma and  above']
```

```python
def education_map(level):
  if level in list1:
    return 'No_education'
  elif level in list2:
    return 'Primary_education'
  elif level in list3:
    return 'Secondary_education'
  elif level in list4:
    return 'Tertiary'
  else:
    return 'unkwon'



```

```python
df3 = df2.copy()
df3['education_level_mapped'] = df2['education_level'].apply(education_map)
```

```python
df3.sample(20)
```

```python
df3['education_level_mapped'].unique()
```

```python
df3.to_csv("eth_sub_final.csv", index = False)
```

```python
def rename_func(df, rename):
    df.urban_rural = df.urban_rural.replace(rename, regex=True)
    return df

df= rename_func(df.copy(), rename.copy())
```

### Aplying Linear Interpolation

```python
df = pd.read_csv("/content/eth_sub_final.csv")
```

```python
df.rename(columns = {'urban/rural':'urban_rural'}, inplace = True)
```

```python
df.drop(columns =['education_level'], axis =1, inplace = True)
```

```python
rename = { 'Urban': 'urban',
          'Rural': 'rural'}
def rename_func(df, rename):
    df.urban_rural = df.urban_rural.replace(rename, regex=True)
    return df

df= rename_func(df.copy(), rename.copy())
```

```python
df.sort_values(by = 'year', inplace = True)
```

```python
df['employed_population'] = df['employed_population'].interpolate(method='linear')
```

```python
df.info()
```

```python
df['unemployed_population'] = df['unemployed_population'].interpolate(method ='linear')
```

```python
df.info()
```

```python
# Group by multiple columns and calculate the sum
grouped_df = df.groupby(['region', 'gender', 'year', 'age group', 'urban_rural', 'education_level_mapped']).agg({
    'employed_population': 'sum',
    'unemployed_population': 'sum'
}).reset_index()
```

```python
# Set up the seaborn plot
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# Bar plot for employed population
sns.barplot(data=grouped_df, x='year', y='employed_population', hue='gender', ci=None)

# labels and title
plt.xlabel('Year')
plt.ylabel('Employed Population')
plt.title('Employed Population Distribution by Year and gender')


plt.show()
```

```python
# Set up the seaborn plot
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# Bar plot for employed population
sns.barplot(data=grouped_df, x='year', y='unemployed_population', hue='gender', ci=None)

# labels and title
plt.xlabel('Year')
plt.ylabel('Employed Population')
plt.title('Unemployed Population Distribution by Year and gender')

plt.show()
```
